```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01"]),
    "Value":[10,13,16,18,24,36,45,56,58,60]
}
```

```python
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
```

```python
plt.plot(df.index,df["Value"],marker="X")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

```python
plt.plot(df.index,df["Value"],marker="o")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

```python
dataseason={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01"]),
    "Value":[10,13,9,18,24,36,30,45,58,48]
}
df=pd.DataFrame(dataseason)
df.set_index("Date",inplace=True)
plt.figure(figsize=(15,6))
```

```python
plt.plot(df.index,df["Value"],marker="o")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Seasonal")
plt.grid(True)
plt.show()
```
