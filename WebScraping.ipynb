```python
import requests
```

```python
link="https://www.goodreads.com/genres/romance"
link_text=requests.get(link).text
print(link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(link_text,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
soup.a
```

```python
soup.find_all('a')
```

```python
(soup.img).attrs['src']
```

```python
(soup.img).attrs['alt']
```

```python
books=soup.find_all('img')
```

```python
data=[]
for book in books:
  if 'alt' in list(book.attrs.keys()):
      data.append(book.attrs['alt'])
```

```python
for i in data:
  print(i)
```
